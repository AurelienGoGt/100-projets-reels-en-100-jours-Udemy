Objectif 1 :

- Creer different cas pour tester les fonctions if/elif/else :
- Utiliser la fonction print(), pour avoir un visuel
- Pour plus de test, creation d'un input pour generer un chiffre different a la fois

```python
print("Exercice 1 !")
number = int(input("Enter a number: "))

if number > 5:
  print("Your number is greater than 5.")
elif number == 5:
  print("Your number is equal to 5.")
elif number == 4:
  print("Your number is equal to 4.")
else:
  print("Your number is less than 5.")

print("\nExercice 2 if a > 5 or/and b < 15!")
a = int(input("Enter a number: "))
b = int(input("Enter another number: "))

if a > 5 or b < 15:
  print("Either one of them is true")
else:
  print("Both conditions are false.")
```

Objectif 2 : 

- Creation d'un outil de comparaison de nombre
- Utilisation de print(), pour visualiser les resultats.
- Utilisation d'input pour choisir les nombres.
- Utilisation de if/elif/else

```python
# Number Comparison Tool

# Step 1: Get two numbers from the user
print("Welcome to the Number Comparison Tool!")
number_1 = float(input("Enter the first number: "))
number_2 = float(input("Enter the second number: "))

# Step 2: Compare the numbers and display results
print("\n--- Comparison Results ----")

if number_1 == number_2:
    print(f"Your numbers are equal : {number_1}")
elif number_1 > number_2:
    print(f"{number_1} is greater than : {number_2}")
else:
    print(f"{number_2} is greater than : {number_1}")

# Step 3: Check if any number is zero and display appropriate message
if number_1 == 0 or number_2 == 0:
    print("\nAt least one number is zero.")
else:
    print("\nBoth numbers are non-zero.")
```
