Objectif 1 :

- Creer different type de variable :

    - String -> Chaine de caractere
    - Integer -> Nombre entier
    - Float -> Nombre a virgule
    - Boolean -> True/False - Vrai/Faux

```python
# Stocker les informations dans des variables :

# Variable de type string (chaine de caractere)
name = "Aurelien"

# Variable de type integer (nombre entier)
age = 35

# Variable de type float (nombre a virgule)
Taille = 1.85

# Variable de type boolean (vrai/faux)
Employe = True

# Afficher les informations stockees dans les variables
print("Nom :", name)
print("Age :", age)
print("Taille :", Taille)
print("Employe :", Employe) 
```
