Objectif 2 : 

- Utilisation d'input pour choisir la duree
- Utilisation de print(), pour visualiser les resultats.
- Utilisation de while pour generer un message avec des emojis

```python
# Timer Script :

import time

# Step 1: input from user :
start = int(input("Enter the number to start the countdown from: "))

# Step 2: While loop for countdown :
print("\n---> Countdown Begins <---")
while start > 0:
  print(start)
  time.sleep(1)
  start -= 1

# Step 3: Print completion message :
print(" Congratulations you succeeded!")
```
