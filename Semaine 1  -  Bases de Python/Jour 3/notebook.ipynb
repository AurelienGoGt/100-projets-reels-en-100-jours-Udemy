Objectif 2 : 

- Creation de la "calculatrice"
- Utilisation d'input pour recuperer les informations
- Utilisation des variables a l'interieur des messages personnalisees pour afficher :
    - Addition
    - Soustraction
    - Multiplication
    - Division

```python
# Calculatrice simple

# Étape 1 : Récupérer les deux nombres saisis par l'utilisateur
number1 = float(input("Entrez le premier nombre : "))
number2 = float(input("Entrez le second nombre : "))

# Étape 2 : Effectuer les opérations arithmétiques
addition = number1 + number2
soustraction = number1 - number2
multiplication = number1 * number2
division = number1 / number2 if number2 != 0 else "Impossible de diviser par zéro"

# Étape 3 : Afficher les résultats
print("\n--- Résultats de la calculatrice ---")
print(f"Addition : {number1} + {number2} = {addition}")
print(f"Soustraction : {number1} - {number2} = {soustraction}")
print(f"Multiplication : {number1} × {number2} = {multiplication}")
print(f"Division : {number1} / {number2} = {division}")

```
